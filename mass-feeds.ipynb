# Adding all the people a twitter user follows to a list

## Goal

We wanted to see what Massachusetts politicians' Twitter feed looks like to get a sense of the media they consume over Twitter. To that end, we compose some lists that have all the people a given politician follows on them.

## Tools
- twitter api
- [python-twitter](https://python-twitter.readthedocs.io/en/latest/index.html), the python wrapper for the twitter api

```python
import twitter, time, json

# Load config from JSON
with open('config.json') as data_file:
    json = json.load(data_file)

# Connect to Twitter API
api = twitter.Api(consumer_key=json['twitter']['consumer_key'],
                  consumer_secret=json['twitter']['consumer_secret'],
                  access_token_key=json['twitter']['access_token_key'],
                  access_token_secret=json['twitter']['access_token_secret'],
                  sleep_on_rate_limit=True)

print api.VerifyCredentials()
```

A list of accounts to get followers from. 

```python
politician = "va_shiva"
list_slug = "shiva-ayyadurai"
```

Get follower user IDs for each follower in the account list. 

```python
follows = api.GetFriendIDs(screen_name=politician)
print follows
```

```python
print len(follows)
```

```python
api.CreateList(politician)
```

```python
chunks = []
follow_strings = []

for i in xrange(0, len(follows), 10):
    chunks.append(follows[i:i + 10])
```

```python
for chunk in chunks:
    api.CreateListsMember(slug="shiva-ayyadurai", user_id=chunk, owner_screen_name="FeedsMass")
    time.sleep(3)
```
